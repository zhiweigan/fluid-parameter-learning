```python
import FluidEnv
import torchaudio
import numpy as np
import skimage.measure
from scipy.signal import hilbert
```

```python
x, sr = torchaudio.load('./data/podo_long.wav')
target_amp_long = np.abs(hilbert(x))
target_mel_long = torchaudio.transforms.MelSpectrogram(sr, 400)(x)
target_amp_long = skimage.measure.block_reduce(target_amp_long, (1, 500), np.max) 
```

```python
import matplotlib.pyplot as plt
```

```python
target_amp_long
```

```python
plt.figure()
plt.plot(target_amp_long[1])
plt.axis('off')
plt.show()
```

```python
plt.imshow(target_mel_long[0,:,:250])
```
